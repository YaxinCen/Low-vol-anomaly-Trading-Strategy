```python
data = pd.read_csv('div_filled.csv', low_memory=False).drop(columns=['datetime date'])
```

```python
# Filter data, get when are the dates used
head, mid, tail = [], [], []
for i1 in range(1964, 2018):
    for i2 in range(2,13):
        mid.append(datetime.datetime(i1, i2, 1)-datetime.timedelta(days = 1))
    mid.append(datetime.datetime(i1+1, 1, 1)-datetime.timedelta(days = 1))

for i1 in range(8, 13):
    head.append(datetime.datetime(1963, i1, 1)-datetime.timedelta(days = 1))
head.append(datetime.datetime(1964, 1, 1)-datetime.timedelta(days = 1))
    
for i1 in range(2, 8):
    tail.append(datetime.datetime(2018,i1,1)-datetime.timedelta(days = 1))

dates_month = head+mid+tail # The index of the data file, in months
```

```python
dates_month
```

```python
data['PRC'] = data['PRC'].fillna(-1)
data['SHROUT'] = data['SHROUT'].fillna(-1)
```

```python
# make sure the dates are aligned to the last day of the month
def match_last_day(date):
    date = datetime.datetime.strptime(date,'%Y/%m/%d')
    if date not in dates_month:
        if date.month<=11:
            return datetime.datetime(date.year, date.month+1, 1) - datetime.timedelta(days=1)
        else:
            return datetime.datetime(date.year+1, 1, 1) - datetime.timedelta(days=1)
    else:
        return date

data['date'] = data['date'].apply(match_last_day)
data
```

```python
# Record dates
cur_tic = data['PERMNO'][0]
start_ind = 0
dates_ind = {}
dates = {}
start_time = time.time()

for i in range(len(data)):
    if data['PERMNO'][i]==cur_tic:
        pass
    else:
        dates_ind.update({cur_tic:[start_ind, i-1]})
        dates.update({cur_tic:[data['date'][start_ind],data['date'][i-1]]})
        start_ind=i
        cur_tic = data['PERMNO'][i]
        
dates_ind.update({cur_tic:[start_ind, i-1]})
dates.update({cur_tic:[data['date'][start_ind],data['date'][i-1]]})
print(time.time()-start_time)

# For some stocks, the shares outstanding and the prices may be missing, 
# if so we need to mark the date when they have actually 

for ticker in dates_ind:
    # Get the start right
    si, ei = dates_ind[ticker][0], dates_ind[ticker][1]
    while(data['PRC'].iloc[si]<0 or data['SHROUT'].iloc[si]<0):
        si+=1
    while(data['PRC'].iloc[ei]<0 or data['SHROUT'].iloc[ei]<0):
        ei-=1
    dates_ind.update({ticker:[si,ei]})
    dates.update({ticker:[data['date'][si],data['date'][ei]]})
    
print(time.time()-start_time)
dates_ind
```

```python
# For a given date, find the stocks that have prices on that day
start_time = time.time()
store = {}

# find_stocks will save the tickers of the stocks available at given date and return a list of them
def find_stocks(target_date):
    stock_lst = []
    for ticker in dates:
        if dates[ticker][0] <= target_date and dates[ticker][1] >= target_date:
            stock_lst.append(ticker)
    return stock_lst


for i in dates_month:
    store.update({i:find_stocks(i)})
    
print(time.time()-start_time)

# The size of each portfolio
for date in store.keys():
    print(len(store[date]))
    
print(time.time()-start_time)
```
